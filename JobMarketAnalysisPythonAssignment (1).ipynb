```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# a) included the sample size & process of data collection
df = pd.read_csv('DataScientistJobVacancies_Indeed.com.csv')
```

```python
# df.head()
df.columns
```

```python
# Removed new lines(\n) in Posted Date (Ex: EmployerActive\n 4 days ago)
df['Posted Date'] = df['Posted Date'].str.replace('\n', '', regex=True)
```

```python
df.to_csv('DataScientistJobVacancies_Indeed.com.csv', index=False)
```

```python
df.head()
```

```python
# removed \n from job type as well
df['Job Type'] = df['Job Type'].str.replace('\n', '', regex=True)
```

```python
# saves the data frame to csv file1
df.to_csv('DataScientistJobVacancies_Indeed.com.csv', index=False)
```

```python
# Part B Visualize the frequency of different job positions listed
job_counts = df['Job Position'].value_counts()
job_counts.head()
```

```python
job_counts.describe()
```

```python
# b) Market Data Visualization
# - Titles of positions available
plt.figure(figsize=(10, 8))
top_job_counts = job_counts.head(10)  # Display only the top 10 job positions
sns.barplot(x=top_job_counts.values, y=top_job_counts.index, palette="viridis")
plt.title("Top 10 Job Positions by Frequency")
plt.xlabel("Number of Postings")
plt.ylabel("Job Position")
plt.show()
```

```python
text = " ".join(job for job in df['Job Position'])
```

```python
from collections import Counter
# Combine all skills into a single list
all_skills = df['Key Skills'].str.split(',').sum()

# Count the frequency of each skill
skill_counts = Counter([skill.strip() for skill in all_skills])
```

```python
# Sample data (replace with your extracted skills)
# Convert to DataFrame for plotting
skills_df = pd.DataFrame(skill_counts.items(), columns=['Skill', 'Count']).sort_values(by='Count', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Skill', data=skills_df, palette="plasma")
plt.title("Key Skills Required for Job Positions")
plt.xlabel("Number of Postings")
plt.ylabel("Skill")
plt.show()
```

```python
# Example of job positions to filter
job_positions = ["Data Scientist", "ML Engineer", "Data Engineer"]

# Filter the data for these job positions
filtered_data = df[df['Job Position'].isin(job_positions)]

# Plotting the salary distribution for each job position
plt.figure(figsize=(10, 6))

# You can plot boxplots to see the distribution of salaries
sns.boxplot(x='Job Position', y='Salary', data=filtered_data, palette="Set3")

plt.title("Salary Distribution by Job Position")
plt.xlabel("Job Position")
plt.ylabel("Salary")
plt.show()
```

```python
# Calculate the average salary for each job position
avg_salaries = filtered_data.groupby('Job Position')['Salary'].mean().reset_index()

# Plotting the average salary
plt.figure(figsize=(10, 6))
sns.barplot(x='Job Position', y='Salary', data=avg_salaries, palette="Set3")

plt.title("Average Salary by Job Position")
plt.xlabel("Job Position")
plt.ylabel("Average Salary")
plt.show()
```
